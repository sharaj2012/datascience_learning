```python
comps = pd.read_csv("data/Competitions.csv")
comps.columns
```

```python
evaluation = ['EvaluationAlgorithmAbbreviation','EvaluationAlgorithmName','EvaluationAlgorithmDescription']
compt = ['Title','EnabledDate','HostSegmentTitle']
df = comps[evaluation+compt].copy()
df['Year'] = pd.to_datetime(df['EnabledDate']).dt.year.values
df['comps']=1
time_select = df.Year >= 2015
compentation_type_select = df.HostSegmentTitle.isin(['Featured','Research'])
```

```python
df[time_select&compentation_type_select]
```

```python
df_new = pd.pivot_table(data=df[time_select&compentation_type_select],values='comps',index=['EvaluationAlgorithmAbbreviation'],columns=['Year'],fill_value=0.0,aggfunc=np.sum,margins=True).sort_values(by=('All'),ascending=False).iloc[1:,:].head(20)
```

```python
df_new
```

```python
metric ='AUC'
metric_select = df['EvaluationAlgorithmAbbreviation'] == metric
df[time_select&compentation_type_select&metric_select][['Title','Year']]
```

```python
counts = df[time_select&compentation_type_select].groupby('EvaluationAlgorithmAbbreviation')
total_comps_per_year = df[time_select&compentation_type_select].groupby('Year').sum()
single_metrics_per_year = counts.sum()[counts.sum().comps==1].groupby('Year').sum()
table = (total_comps_per_year.rename(columns={'comps': 'n_comps'}).join(single_metrics_per_year / total_comps_per_year).rename(columns={'comps': 'pct_comps'}))
print(table)
```
